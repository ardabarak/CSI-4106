For my data processing I have also worked with my data, which I will show below.

In order I have removed the attributes I found not useful earlier (I will show before and after below)

```python
#my dataset visually before dropping the attributes
datasetWine.head()
```

```python
#dropping the attributes
datasetWine = datasetWine.drop('Id', axis=1)
datasetWine = datasetWine.drop('pH', axis=1)
datasetWine = datasetWine.drop('free sulfur dioxide', axis=1)
datasetWine = datasetWine.drop('residual sugar', axis=1)
#my dataset after dropping the mentioned attributes
datasetWine.head()
```

I now will be cleaning values for some of the attributes which I found to be misleading from their boxplot visuals (exceeding the maximum values)

```python
# = the values > maximum to median of the dataset for each attribute in this code block and the ones below
attribute1 = 'volatile acidity'

#median value
median_value = datasetWine[attribute1].median()
#maximum value
max_value = datasetWine[attribute1].max() #should be max 1 for volatile acidity

#values exceeding the max are now going to be equal to the median
datasetWine[attribute1] = datasetWine[attribute1].apply(lambda x: median_value if x > max_value else x)
datasetWine[attribute1].describe() #seeing the new changes
```

```python
attribute2 = 'citric acid'

median_value = datasetWine[attribute2].median()
max_value = datasetWine[attribute2].max() #should be max around 0.75to0.80, values above are not usual

datasetWine[attribute2] = datasetWine[attribute2].apply(lambda x: median_value if x > max_value else x)
datasetWine[attribute2].describe()
```

```python
#Instead of replacing chlorides values above the maximum amount,
# I will be removing the values above 0.3(above the max of the boxplot)
#  as chlorides above 0.3 might lead to contamination in wine and that could mislead our data later on

attribute3 = 'chlorides'

#cleaning data from the rows where the chloride value exceeds 0.3
datasetWine = datasetWine[datasetWine[attribute3] <= 0.3]

datasetWine.describe()
```


I'll now clean the dataset from the lines with the exact same values (duplicates)


```python
#doing this after the above operations as some values might've not previously been the same but might be now (after value changing operations)
datasetWine = datasetWine.drop_duplicates()

datasetWine.describe()
```

Now I will show the before and after of attribute distributions before applying the box cox transformations below.



```python
#before
showHist = ['fixed acidity', 'total sulfur dioxide']
datasetWine[showHist].hist(bins=20, figsize=(15, 5))
plt.show()
```

```python
#transforming data
powerTransform = PowerTransformer(method='box-cox', standardize=False)  #boxcoxing attributes>=0

boxCoxedFA = powerTransform.fit_transform(datasetWine[['fixed acidity']])
boxCoxedTSD = powerTransform.fit_transform(datasetWine[['total sulfur dioxide']])

datasetWine['fixed acidity'] = boxCoxedFA
datasetWine['total sulfur dioxide'] = boxCoxedTSD
# https://scikit-learn.org/stable/modules/preprocessing.html

```

```python
#after
showHist = ['fixed acidity', 'total sulfur dioxide']
datasetWine[showHist].hist(bins=20, figsize=(15, 5))
plt.show()
```
